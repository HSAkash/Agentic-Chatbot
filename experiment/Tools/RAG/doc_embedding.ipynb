## Doc file to pdf

```python

input_doc = "/home/yato/ML/Agentic-Chatbot/data/unstructured_docs/test_doc/Sample-doc-file-1000kb.doc"
output_docx = "/home/yato/ML/Agentic-Chatbot/data/unstructured_docs/test_doc/Sample-doc-file-1000kb.pdf"


from spire import doc
from spire.doc.common import *
        
# Create a Document object
document = doc.Document()
# Load a Word DOCX file
document.LoadFromFile(input_doc)
# Or load a Word DOC file
#document.LoadFromFile("Sample.doc")

# Save the file to a PDF file
document.SaveToFile(output_docx, doc.FileFormat.PDF)
document.Close()
```
